# FIRST NOTEBOOK  
## Prompt Engineering Quiz  

---

### Question 1  
**What is prompting?**  

**Answer:**  
Prompting is the process of giving clear instructions or inputs to an AI model that guide its response. The way you phrase the request determines the quality, accuracy, and style of the output.  

---

### Question 2  
**In the following prompt, which part describes the primary task?**  

*Prompt:*  
“Explain how car batteries work using non-technical language that someone without any mechanical knowledge could understand and interpret.”  

**Answer:**  
The primary task is: *“Explain how car batteries work using non-technical language.”*  
Everything else adds context about the audience and tone.  

---

### Question 3  
**When you are using a generative AI tool to create a healthy three-course menu that requires a certain number of calories, which prompting element will ensure the menu is more practical?**  

**Answer:**  
Adding **constraints** (such as calorie count, number of courses, or healthy options) ensures practicality.  
Providing **examples** (like a sample menu) through one-shot prompting also helps the AI align with real-world requirements.  

---

### Question 4  
**Prompting with audio to receive a visual representation is called what?**  

**Answer:**  
This is called **multimodal prompting**, since it uses one type of input (audio) to generate another type of output (visual).  

---

### Question 5  
**What does a human-in-the-loop approach to prompting emphasize?**  

**Answer:**  
It emphasizes that humans remain actively involved in checking, refining, and validating AI outputs.  
The human ensures results are accurate, ethical, and aligned with the intended goal instead of relying blindly on the AI.  

---

### Question 6  
**When the initial prompt “Create a dataset” is too broad, what should you do to refine it?**  

**Answer:**  
Refine it by adding **specific context** — for example:  
- What topic the dataset should cover.  
- How many records are required.  
- In what format (CSV, JSON, or table).  

The more detail given, the more useful and structured the dataset will be.

# DAILY PRODUCTIVITY & MOTIVATION CHATBOT

---

## Description  
This chatbot is designed as a **simple daily assistant** that helps users plan their day and stay motivated.  
It:  
1. Greets the user and asks for their **name and current mood**.  
2. Provides a **personalized motivational response** based on mood.  
3. Lets the user **list 2–3 important tasks**, then organizes them into a small plan.  
4. Ends with a motivational reminder to encourage consistency.  

The chatbot is lightweight, interactive, and runs directly in a Python notebook.  
It follows the principles of **prompting and conversational flow** learned in earlier exercises.

---

## Code Implementation

```python
# Chatbot: Daily Productivity & Motivation Assistant

def chatbot():
    print("Hello! I’m your Daily Productivity & Motivation Chatbot.")
    
    # Greeting & context
    name = input("What’s your name? ")
    mood = input(f"Nice to meet you, {name}! How are you feeling today? (happy/sad/tired/motivated) ")
    
    # Motivational response
    if mood.lower() in ["sad", "tired"]:
        print("Remember: Small steps every day create big changes. You’ve got this!")
    elif mood.lower() == "motivated":
        print("Love the energy! Let’s make today productive.")
    else:
        print("Glad to hear that! Let’s keep up the positive vibes.")
    
    # Task planner
    print("\nLet’s plan your day. Tell me 2–3 important tasks.")
    tasks = []
    for i in range(3):
        task = input(f"Task {i+1}: ")
        if task.strip() != "":
            tasks.append(task)
    
    print("\nHere’s your simple day plan:")
    for i, t in enumerate(tasks, 1):
        print(f"{i}. {t}")
    
    # Closing
    print(f"\nStay focused, stay consistent. Wishing you a productive day ahead, {name}!")

# Run chatbot
chatbot()
```
